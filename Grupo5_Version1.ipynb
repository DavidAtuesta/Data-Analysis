```python
#Quitar de la correlación los datos no númericos
df_numerico = df_grupo_5.select_dtypes(include=[np.number])
df_numerico
```

```python
correlacion = df_numerico.corr()
correlacion
```

```python
#Encontrar las correlaciones superiores a 0,5
correlaciones_significativas = correlacion[(correlacion.abs() > 0.5) & (correlacion != 1.0)]
correlaciones_significativas
```

```python
#Mostrar las variables correlacionadas
for columna in correlaciones_significativas.columns:
    correlaciones = correlaciones_significativas[columna].dropna()
    if not correlaciones.empty:
        print(f"Variables correlacionadas con {columna}:")
        for indice, valor in correlaciones.items():
            print(f"  {indice}: {valor}")
```

```python
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
def scatter_plot_with_trendline(x, y, xlabel, ylabel, title):
    plt.figure(figsize=(10, 6))
    plt.scatter(x, y, color='blue')

    # Ajustar la línea de tendencia
    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    plt.plot(x, slope*x + intercept, color='red', linestyle='--')

    # Añadir la ecuación de la línea y el coeficiente de determinación
    equation = f'y = {slope:.2f}x + {intercept:.2f}'
    r_squared = f'$R^2$ = {r_value**2:.2f}'
    plt.text(0.05, 0.95, equation, transform=plt.gca().transAxes, fontsize=12, verticalalignment='top', fontweight='bold')
    plt.text(0.05, 0.90, r_squared, transform=plt.gca().transAxes, fontsize=12, verticalalignment='top', fontweight='bold')

    # Configurar el gráfico
    plt.title(title, fontweight='bold')
    plt.xlabel(xlabel, fontweight='bold')
    plt.ylabel(ylabel, fontweight='bold')
    plt.xticks(fontweight='bold')
    plt.yticks(fontweight='bold')
    plt.grid(False)
    plt.show()
```

```python
# Total_Bovinos_(Unidad) vs Producción_CH4_Anual_(Ton)
scatter_plot_with_trendline(df_grupo_5['Total_Bovinos_(Unidad)'], df_grupo_5['Producción_CH4_Anual_(Ton)'],
                            'Total Bovinos (Unidad)', 'Producción CH4 Anual (Ton)',
                            'Total Bovinos (Unidad) vs Producción CH4 Anual (Ton)')

# Producción_Leche_Anual_(M3) vs Producción_CH4_Anual_(Ton)
scatter_plot_with_trendline(df_grupo_5['Producción_Leche_Anual_(M3)'], df_grupo_5['Producción_CH4_Anual_(Ton)'],
                            'Producción Leche Anual (M3)', 'Producción CH4 Anual (Ton)',
                            'Producción Leche Anual (M3) vs Producción CH4 Anual (Ton)')

# Hembras_lecheras_(Unidad) vs Producción_Leche_Anual_(M3)
scatter_plot_with_trendline(df_grupo_5['Hembras_lecheras_(Unidad)'], df_grupo_5['Producción_Leche_Anual_(M3)'],
                            'Hembras lecheras (Unidad)', 'Producción Leche Anual (M3)',
                            'Hembras lecheras (Unidad) vs Producción Leche Anual (M3)')

# Total_Bovinos_(Unidad) vs Hembras_lecheras_(Unidad)
scatter_plot_with_trendline(df_grupo_5['Total_Bovinos_(Unidad)'], df_grupo_5['Hembras_lecheras_(Unidad)'],
                            'Total Bovinos (Unidad)', 'Hembras lecheras (Unidad)',
                            'Total Bovinos (Unidad) vs Hembras lecheras (Unidad)')

# Pastos_(Ha) vs Total_Bovinos_(Unidad)
scatter_plot_with_trendline(df_grupo_5['Pastos_(Ha)'], df_grupo_5['Total_Bovinos_(Unidad)'],
                            'Pastos (Ha)', 'Total Bovinos (Unidad)',
                            'Pastos (Ha) vs Total Bovinos (Unidad)')

# Pastos_(Ha) vs Producción_CH4_Anual_(Ton)
scatter_plot_with_trendline(df_grupo_5['Pastos_(Ha)'], df_grupo_5['Producción_CH4_Anual_(Ton)'],
                            'Pastos (Ha)', 'Producción CH4 Anual (Ton)',
                            'Pastos (Ha) vs Producción CH4 Anual (Ton)')
```

```python
def plot_line_chart(x, y1, y2, xlabel, ylabel1, ylabel2, title, color1, color2):
    plt.figure(figsize=(16, 8))
    plt.plot(x, y1, color=color1, label=ylabel1, linewidth=2)
    plt.plot(x, y2, color=color2, label=ylabel2, linewidth=2)
    plt.title(title, fontweight='bold')
    plt.xlabel(xlabel, fontweight='bold')
    plt.ylabel('Valores', fontweight='bold')
    plt.xticks(rotation=45, ha='right', fontweight='bold')
    plt.yticks(fontweight='bold')
    plt.legend()
    plt.grid(False)
    plt.tight_layout()
    plt.show()

# Producción_Leche_Anual_(M3) vs Aumento_leche_Acacia_(M3)
plot_line_chart(df_grupo_5['Municipios'], df_grupo_5['Producción_Leche_Anual_(M3)'], df_grupo_5['Aumento_leche_Acacia_(M3)'],
                'Municipios', 'Producción Leche Anual (M3)', 'Aumento leche Acacia (M3)',
                'Producción Leche Anual (M3) vs Aumento leche Acacia (M3)', 'gold', 'green')

# Producción_Leche_Anual_(M3) vs Aumento_leche_Baccharis_(M3)
plot_line_chart(df_grupo_5['Municipios'], df_grupo_5['Producción_Leche_Anual_(M3)'], df_grupo_5['Aumento_leche_Baccharis_(M3)'],
                'Municipios', 'Producción Leche Anual (M3)', 'Aumento leche Baccharis (M3)',
                'Producción Leche Anual (M3) vs Aumento leche Baccharis (M3)', 'gold', 'purple')

# Producción_Leche_Anual_(M3) vs Aumento_leche_Sambucus_(M3)
plot_line_chart(df_grupo_5['Municipios'], df_grupo_5['Producción_Leche_Anual_(M3)'], df_grupo_5['Aumento_leche_Sambucus_(M3)'],
                'Municipios', 'Producción Leche Anual (M3)', 'Aumento leche Sambucus (M3)',
                'Producción Leche Anual (M3) vs Aumento leche Sambucus (M3)', 'gold', 'blue')
```

```python
df_proyecciones = pd.read_excel('PROYECCIONES_2025_2028.xlsx')
# Asignación de nombres a las columnas
df_proyecciones.columns = [
    'Municipio',
    'Poblacion_2021',
    'Metano_2021',
    'Hembras_2021',
    'Leche_2021',
    'Poblacion_2025',
    'Metano_2025',
    'Hembras_2025',
    'Leche_2025',
    'Poblacion_2028',
    'Metano_2028',
    'Hembras_2028',
    'Leche_2028'
]
df_proyecciones = df_proyecciones.drop(df_proyecciones.index[0]) # Borrar una fila innecesaria
df_proyecciones
```

```python
df_proyecciones.info()
```

```python
#PROYECCIÓN DE PRODUCCIÓN DE METANO EN FUNCIÓN DE LA POBLACIÓN PROYECTADA PARA 2025 Y 2028

# Definir la ecuación
def calcular_ch4(poblacion):
  return 0.12 * poblacion + 4.01

# Calcular las producciones de CH4 proyectadas
df_proyecciones['CH4_2025'] = df_proyecciones['Poblacion_2025'].apply(calcular_ch4)
df_proyecciones['CH4_2028'] = df_proyecciones['Poblacion_2028'].apply(calcular_ch4)

# Graficar los resultados
plt.figure(figsize=(10, 6))
plt.plot(df_proyecciones['Municipio'], df_proyecciones['Metano_2021'], label='CH4 2021', marker='o')
plt.plot(df_proyecciones['Municipio'], df_proyecciones['CH4_2025'], label='CH4 2025', marker='o')
plt.plot(df_proyecciones['Municipio'], df_proyecciones['CH4_2028'], label='CH4 2028', marker='o')
plt.xlabel('Municipio', fontweight='bold')
plt.ylabel('Producción de CH4 (Ton/Año)', fontweight='bold')
plt.title('Proyección de Producción de CH4 en función de la población bovina proyectada', fontweight='bold')
plt.xticks(rotation=45, ha='right', fontweight='bold')
plt.yticks(fontweight='bold')
plt.legend(prop={'weight': 'bold'})
plt.tight_layout()
plt.show()
```

```python
#PROYECCIÓN DE HEMBRAS LECHERAS EN FUNCIÓN DE LA POBLACIÓN PROYECTADA PARA 2025 Y 2028

# Definir la ecuación
def calcular_hembras(poblacion):
  return 0.53 * poblacion - 734.21

# Calcular las hembras proyectadas
df_proyecciones['Hembras_2025'] = df_proyecciones['Poblacion_2025'].apply(calcular_hembras)
df_proyecciones['Hembras_2028'] = df_proyecciones['Poblacion_2028'].apply(calcular_hembras)

# Graficar los resultados
plt.figure(figsize=(10, 6))
plt.plot(df_proyecciones['Municipio'], df_proyecciones['Hembras_2021'], label='Hembras 2021', marker='o')
plt.plot(df_proyecciones['Municipio'], df_proyecciones['Hembras_2025'], label='Hembras 2025', marker='o')
plt.plot(df_proyecciones['Municipio'], df_proyecciones['Hembras_2028'], label='Hembras 2028', marker='o')
plt.xlabel('Municipio', fontweight='bold')
plt.ylabel('Número de Hembras', fontweight='bold')
plt.title('Proyección del número de hembras en función de la población bovina proyectada', fontweight='bold')
plt.xticks(rotation=45, ha='right', fontweight='bold')
plt.yticks(fontweight='bold')
plt.legend(prop={'weight': 'bold'})
plt.tight_layout()
plt.show()
```

```python
#PROYECCIÓN DE PRODUCCIÓN  DE METANO EN FUNCIÓN DE LA PRODUCCIÓN DE LECHE PROYECTADA PARA 2025 Y 2028

# Definir la ecuación
def calcular_ch4_leche(leche):
  return 0.03 * leche + 628.54

# Calcular la producción de CH4 proyectado
df_proyecciones['Metano_2025'] = df_proyecciones['Leche_2025'].apply(calcular_ch4_leche)
df_proyecciones['Metano_2028'] = df_proyecciones['Leche_2028'].apply(calcular_ch4_leche)

# Graficar los resultados
plt.figure(figsize=(10, 6))
plt.plot(df_proyecciones['Municipio'], df_proyecciones['Metano_2021'], label='CH4 2021', marker='o')
plt.plot(df_proyecciones['Municipio'], df_proyecciones['Metano_2025'], label='CH4 2025', marker='o')
plt.plot(df_proyecciones['Municipio'], df_proyecciones['Metano_2028'], label='CH4 2028', marker='o')
plt.xlabel('Municipio', fontweight='bold')
plt.ylabel('Producción de CH4 (Ton/Año)', fontweight='bold')
plt.title('Proyección de Producción de CH4 en función de la producción de leche proyectada', fontweight='bold')
plt.xticks(rotation=45, ha='right', fontweight='bold')
plt.yticks(fontweight='bold')
plt.legend(prop={'weight': 'bold'})
plt.tight_layout()
plt.show()
```

```python
#PROYECCIÓN DE LA PRODUCCIÓN DE LECHE EN FUNCIÓN DE LA POBLACIÓN DE HEMBRAS PROYECTADA PARA 2025 Y 2028

# Definir la ecuación
def calcular_leche(hembras):
  return 6.04 * hembras - 453.2

# Calcular la producción de leche proyectada
df_proyecciones['Leche_2025'] = df_proyecciones['Hembras_2025'].apply(calcular_leche)
df_proyecciones['Leche_2028'] = df_proyecciones['Hembras_2028'].apply(calcular_leche)

# Graficar los resultados
plt.figure(figsize=(10, 6))
plt.plot(df_proyecciones['Municipio'], df_proyecciones['Hembras_2021'], label='Leche 2021', marker='o')
plt.plot(df_proyecciones['Municipio'], df_proyecciones['Hembras_2025'], label='Leche 2025', marker='o')
plt.plot(df_proyecciones['Municipio'], df_proyecciones['Hembras_2028'], label='Leche 2028', marker='o')
plt.xlabel('Municipio', fontweight='bold')
plt.ylabel('Producción de Leche (M3/Año)', fontweight='bold')
plt.title('Proyección de producción de leche en función de la población de hembras proyectada', fontweight='bold')
plt.xticks(rotation=45, ha='right', fontweight='bold')
plt.yticks(fontweight='bold')
plt.legend(prop={'weight': 'bold'})
plt.tight_layout()
plt.show()
```
